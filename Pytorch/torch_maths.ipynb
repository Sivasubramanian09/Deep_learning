```python
#Dot Products nd Multiplication
```

```python
x = torch.tensor([3,4,5],dtype= torch.float64)
y = torch.tensor([6,7,8],dtype= torch.float64)
```

```python
x.mul(y)
```

```python
x = torch.tensor([[3,4,5],[1,2,3]],dtype= torch.float64)
y = torch.tensor([[6,7,8],[3,4,5],[4,5,6]],dtype= torch.float64)
```

```python
torch.matmul(x,y)
```

```python
torch.mm(x,y)
```
